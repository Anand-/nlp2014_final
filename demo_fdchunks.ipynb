```python
data_dir = path.join(path.curdir, 'data')
pickle_dir = path.join(path.curdir, 'pickles')
```

```python
hashes={
    "Voter ID": "voterID_clean.pkl",
    "Keystone XL": "Keystonexl.pkl",
    "Too Many Cooks": "Cooks.pkl",
    "Ferguson": "Ferguson.pkl",
    "Gamergate": "GamerGate.pkl"
}
```

```python
def get_results(corpus, chunk_start=0, chunk_end=100, min_matches=3):
    result = chunk_approach(
        path.join(pickle_dir, hashes[corpus]),
        chunk_start=int(chunk_start),
        chunk_end=int(chunk_end),
        min_matches=int(min_matches)
    )
    display_results(result)
```

```python
w=interact(get_results,
         corpus=hashes.keys(),
         chunk_start=(0,100),
         chunk_end=(0,200),
         min_matches=(0,10)
         )
```

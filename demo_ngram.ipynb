```python
from os import path
import cPickle as pickle

from approach_ngrams import ngram_approach
from display_util import display_results

from IPython.html.widgets import interact
```

```python
%matplotlib inline
```

```python
data_dir = path.join(path.curdir, 'data')
pickle_dir = path.join(path.curdir, 'pickles')
```

```python
hashes={
    "Voter ID": "voterID_clean.pkl",
    "Keystone XL": "Keystonexl.pkl",
    "Too Many Cooks": "Cooks.pkl",
    "Ferguson": "Ferguson.pkl",
    "Gamergate": "GamerGate.pkl"
}
```

```python
def get_results(corpus, min_matches=3):
    result = ngram_approach(
        path.join(pickle_dir, hashes[corpus]),
        min_matches=min_matches
    )
    display_results(result)
```

```python
interact(get_results,
         corpus=hashes.keys(),
         min_matches=(0,10)
         )
```

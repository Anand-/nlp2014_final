```python
good_chunks=chunking_util.get_best_chunks(df)
```

```python
chunk_map = {}
for k,v in good_chunks.iteritems():
    chunk_map[k]=k
    for val in v:
        chunk_map[val]=k
```

```python
def apply_map(row, chunk_map):
    return set([chunk_map[c.lower()] for c in row if c.lower() in chunk_map])
```

```python
df_overlap=grouping_util.get_overlaps(chunk_map, df)
```

```python
df['common_chunks']=grouping_util.get_keyphrase_columns(chunk_map, df, make_keyphrase_cols=False)
```

```python
groups=grouping_util.get_groups(df_overlap, doc_percents=False,set_filters=df.common_chunks, min_matches=2)
```
